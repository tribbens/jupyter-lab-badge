```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Create Binomial Distribution

binomial_data = np.random.binomial(100, 0.50, size=100)
plt.hist(binomial_data, bins=10)
plt.show()
```

```python
# Create Uniform Distribution

uniform_data = np.random.uniform(0, 10, size=20)
plt.hist(uniform_data, bins=10)
plt.show()
```

```python
# Create Normal Distribution

uniform_data = np.random.normal(50, 5, size=100)
plt.hist(uniform_data, bins=20)
plt.show()
```
